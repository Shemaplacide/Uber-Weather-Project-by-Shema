```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

df['distance'] = df.apply(lambda row: haversine(
    row['pickup_latitude'], row['pickup_longitude'],
    row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
```

```python
import pandas as pd

# Load Uber data
uber_df = pd.read_csv(r"D:\.kaggle\archive\uber.csv")

# Show columns to understand dataset structure
print("Uber columns:", uber_df.columns.tolist())

# Rename columns if they exist (adjust this map if your actual column names differ)
rename_map = {}
if 'Pickup_datetime' in uber_df.columns:
    rename_map['Pickup_datetime'] = 'pickup_datetime'
if 'Dropoff_datetime' in uber_df.columns:
    rename_map['Dropoff_datetime'] = 'dropoff_datetime'

if rename_map:
    uber_df.rename(columns=rename_map, inplace=True)
    print(f"Renamed columns: {rename_map}")

# Convert pickup_datetime if available
if 'pickup_datetime' in uber_df.columns:
    uber_df['pickup_datetime'] = pd.to_datetime(uber_df['pickup_datetime'])
else:
    print("Warning: 'pickup_datetime' column not found.")

# Convert dropoff_datetime if available
if 'dropoff_datetime' in uber_df.columns:
    uber_df['dropoff_datetime'] = pd.to_datetime(uber_df['dropoff_datetime'])
else:
    print("Warning: 'dropoff_datetime' column not found.")

# Prepare date column for merging if pickup_datetime exists
if 'pickup_datetime' in uber_df.columns:
    uber_df['date'] = uber_df['pickup_datetime'].dt.date
else:
    print("Cannot create 'date' column without 'pickup_datetime'. Exiting.")
    exit(1)

# Load weather data
weather_df = pd.read_csv(r"D:\.kaggle\archive\weather_data.csv")

# Prepare date column in weather data
weather_df['date'] = pd.to_datetime(weather_df['date']).dt.date

# Merge on date
combined_df = pd.merge(uber_df, weather_df, on='date', how='left')

# Calculate trip_duration if both datetime columns exist
if 'pickup_datetime' in combined_df.columns and 'dropoff_datetime' in combined_df.columns:
    combined_df['trip_duration'] = (combined_df['dropoff_datetime'] - combined_df['pickup_datetime']).dt.total_seconds() / 60
else:
    print("Skipping trip_duration calculation due to missing datetime columns.")

# Save merged dataset
combined_df.to_csv(r"D:\.kaggle\archive\uber_with_weather.csv", index=False)

print("✅ Merged data saved to D:\\.kaggle\\archive\\uber_with_weather.csv")
```
